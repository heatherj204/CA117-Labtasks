## MP3Track part 2 [10 marks]

- In module `mp3track_v2_122.py` extend the `MP3Track` class such
  that one or more artists can be associated with a track.  
- When your class is correctly implemented, running the following program
should produce the given output (artists are printed in the order they
are added to a track).  

````python
# TODO: Define MP3Track
class MP3Track(object):
    def __init__(self, title, duration, artist) -> None:
        self.title = title
        self.duration = duration
        self.artist = artist
    def get_artist(self):
        return f'{", ".join(self.artist)}'
    def __str__(self) -> str:
        return (f'Title: {self.title}\nBy: {self.get_artist()} \nDuration: {self.duration}')
````

````python
t1 = MP3Track('Fools Gold', 604, ['The Stone Roses'])
t2 = MP3Track('Shallow', 197, ['Lady Gaga', 'Bradley Cooper'])
t3 = MP3Track('Telephone', 220, ['Beyonce', 'Lady Gaga'])

print(t1)
print(t2)
print(t3)
````
